## 3. Create a New Google Sheet

This cell creates a new Google Sheet for your blog content calendar and prints its URL.

```python
# Create the sheet
sheet_title = 'Ziyada Blog Content Calendar'
worksheet = client.create(sheet_title)
worksheet.share('your-google-account@gmail.com', perm_type='user', role='writer')  # Update with your email

# Open the sheet and set up columns
sheet = client.open(sheet_title).sheet1
columns = [
    'Date', 'Title', 'Language', 'Status', 'Telegram Thread', 'Draft Link', 'Email Drafted',
    'Posted Link', 'Notes', 'Author', 'SEO Keywords', 'Audience', 'CTA', 'Approval Date', 'Scheduled Date'
]
sheet.insert_row(columns, 1)
print(f'Sheet created: https://docs.google.com/spreadsheets/d/{worksheet.id}')
```
